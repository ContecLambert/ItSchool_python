### titanic.csv 파일을 이용해서 아래 문제1) ~ 문제4) 풀어 보세요.

'''
PassengerId(승객번호),Survived(생존여부:1-생존, 0-사망),Pclass(승선권 클래스:1-ㅣ1st, 2-2nd, 3-3rd),Name(승객이름),Sex(승객 성별),Age(승객 나이),SibSp(동반한 형제자매, 배우자 수),Parch(동반한 부모, 자식 수),Ticket(티켓의 고유 넘버),Fare(티켓의 요금),Cabin(객실 번호),Embarked(승선한 항구명: C-Cherbourg, !-Queenstown, S-Southampton)
'''



```python
import pandas as pd
titanic = pd.read_csv('titanic.csv') #타이타닉호 데이터
titanic
```

#### 문제1)  남자(male) 이면서 Pclass 가 1인 승객은 몇 명인가요?

```python
a = titanic.loc[(titanic['Sex']=='male') & (titanic['Pclass']==1), 'PassengerId'].count()
a
```

####  문제2) 남자이면서 Pclass 가 1이고 Fare가 50 이상인 승객은 몇 명인가요?

```python
titanic.loc[(titanic['Sex']=='male') & (titanic['Pclass']==1)&(titanic['Fare'] >=50), 'PassengerId'].count()
```

#### 문제3) 새로운 열(sex_code)을 만들어서 남자인 경우는 코드 0, 여자인 경우는 1로 저장

```python
a = titanic['Sex']=='female'
a = a.astype(int)
titanic['sex_code'] = a
titanic
```

#### 문제4) 성별 생존자 수의 합(남자 생존자 수의 합, 여자 생존자수의 합) - 생존자는 Survived 의 값이 1인 것.

```python
a = titanic.loc[(titanic['Sex']=='male') & (titanic['Survived']==1), 'PassengerId'].count()
b = titanic.loc[(titanic['Sex']=='female') & (titanic['Survived']==1), 'PassengerId'].count()
print('남자생존자수 :',a)
print('여자생존자수 :',b)
```

#### 문제7) 각 영화 별 평균 평점을 movieid, title, 평균 평점 순으로 출력 하세요.

```python
a = df_r.groupby('movieId')['rating'].mean()
a = a.to_frame()
a.columns = ['average']
df = pd.merge(df_m,a, left_on= 'movieId', right_index=True)
df[['movieId','title','average']]
```

#### 문제8) 평균 평점이 가장 높은 영화의 제목을 출력(동률인 경우 모두 출력) 하세요.

```python
df.loc[df["average"]==df['average'].max(), 'title']
```

#### 문제10) 2015년도에 평가된 모든 Romance 영화의 평균 평점을 구하세요.


```python
import time
a = pd.merge(df_m,df_r,left_on ="movieId",right_on="movieId")
a = a.loc[a['genres'].str.contains('Romance')]
a
```

```python
start = time.mktime(time.strptime("2015-01-01 00:00:00", "%Y-%m-%d %H:%M:%S"))
end = time.mktime(time.strptime("2016-01-01 00:00:00", "%Y-%m-%d %H:%M:%S"))
print(a.loc[(a["timestamp"]>=start) & (a["timestamp"]<end),"rating"].mean())
```
