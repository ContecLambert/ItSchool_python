### ARIMA 모델

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
```

```python
# (AR = 2, 차분 = 1 , MA = 2) 파라미터로 ARIMA 모델을 학습합니다.
model = ARIMA(confirmed.confirmed.values, order=(2,1,2))
model_fit = model.fit(trend='c', full_output=True, disp=True)
print(model_fit.summary())
```

```python
fig = model_fit.plot_predict()
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
```

```python
forecast_data = model_fit.forecast(steps=5) 

test_file_path = 'time_test.csv'
confirmed_test_df = pd.read_csv(test_file_path, names=['ds', 'y'])

pred_y = forecast_data[0].tolist() 
test_y = confirmed_test_df.y.values 
pred_y_lower = [] 
pred_y_upper = []
for lower_upper in forecast_data[2]:
    lower = lower_upper[0]
    upper = lower_upper[1]
    pred_y_lower.append(lower)
    pred_y_upper.append(upper)
```

```python
plt.plot(pred_y, color="gold") # 모델이 예상한 확진자 그래프입니다.
plt.plot(pred_y_lower, color="red") # 모델이 예상한 최소 확진자 그래프입니다.
plt.plot(pred_y_upper, color="blue") # 모델이 예상한 최대 확진자 그래프입니다.
plt.plot(test_y, color="green") # 실제 확진자 그래프입니다.
```

```python
plt.plot(pred_y, color="gold") # 모델이 예상한 확진자 그래프입니다.
plt.plot(test_y, color="green") # 실제 확진자 그래프입니다.
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

rmse = sqrt(mean_squared_error(pred_y, test_y))
print(rmse)
```

### Additive 모델

```python
from fbprophet import Prophet
```

```python
confirmed
```

```python
# prophet을 사용하기 위해서는 다음과 같이 피처의 이름을 변경해야 합니다 : 'ds', 'y'
confirmed.rename(columns={confirmed.columns[0]:'ds',
                         confirmed.columns[1]:'y'},inplace=True)
```

```python
prophet = Prophet(seasonality_mode='multiplicative', 
                  yearly_seasonality=True,
                  weekly_seasonality=True, daily_seasonality=True,
                  changepoint_prior_scale=0.5)
prophet.fit(confirmed)
```

```python
# 5일을 내다보며 예측합니다.
future_data = prophet.make_future_dataframe(periods=5, freq='d')
forecast_data = prophet.predict(future_data)
```

```python
forecast_data.tail(5)
```

```python
forecast_data[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(5)
```

```python
fig1 = prophet.plot(forecast_data)
```

```python
fig2 = prophet.plot_components(forecast_data)
```

```python
y = confirmed.y.values[5:] # 첫 5일을 제외한 실제 확진자 데이터입니다.
y_pred = forecast_data.yhat.values[5:-5] # 첫 5일, 마지막 5일을 제외한 예측 확진자 데이터입니다.
```

```python
rmse = sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)
print(rmse)
print(r2)
```

```python
confirmed_test = pd.read_csv('time_test.csv', names=['ds', 'y'])

pred_y = forecast_data.yhat.values[-5:] # 마지막 5일의 예측 데이터입니다. (2020-06-01 ~ 2020-06-05)
test_y = confirmed_test.y.values # 실제 5일 확진자 데이터입니다. (2020-06-01 ~ 2020-06-05)
pred_y_lower = forecast_data.yhat_lower.values[-5:] # 마지막 5일의 예측 데이터의 최소값입니다.
pred_y_upper = forecast_data.yhat_upper.values[-5:] # 마지막 5일의 예측 데이터의 최대값입니다.
```

```python
plt.plot(pred_y, color="gold") # 모델이 예상한 확진자 그래프입니다.
plt.plot(pred_y_lower, color="red") # 모델이 예상한 최소 확진자 그래프입니다.
plt.plot(pred_y_upper, color="blue") # 모델이 예상한 최대 확진자 그래프입니다.
plt.plot(test_y, color="green") # 실제 확진자 그래프입니다.
```

```python
plt.plot(pred_y, color="red", ls="--", marker="o", ms=5) # 모델이 예상한 확진자 그래프입니다.
plt.plot(test_y, color="blue") # 실제 확진자 그래프입니다.
```

```python
rmse = sqrt(mean_squared_error(pred_y, test_y))
print(rmse)
```
