```python
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
# mpl.rc('font', family = 'Malgun Gothic')
```

```python
import numpy as np

x = np.arange(0,6.)   # log n
y = 2**x # n
# z = x*y   # n log n
# a = y**2 #n^2
# b = y**3 # n^3
# c = 2**y # 2^n
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label ='log n', color = 'black')
# ax.plot(x,y, label ='n', color = 'red')
# ax.plot(x, x*y, label ='n log n', color = 'purple')
# ax.plot(x, y**2, label ='n**2', color = 'blue')
# ax.plot(x, y**3, label ='n**3', color = 'green')
# ax.plot(x, 2**y, label ='2**n', color = 'orange')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label ='log n', color = 'black')
ax.plot(x,y, label ='n', color = 'red')
# ax.plot(x, x*y, label ='n log n', color = 'purple')
# ax.plot(x, y**2, label ='n**2', color = 'blue')
# ax.plot(x, y**3, label ='n**3', color = 'green')
# ax.plot(x, 2**y, label ='2**n', color = 'orange')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label ='log n', color = 'black')
ax.plot(x,y, label ='n', color = 'red')
ax.plot(x, x*y, label ='n log n', color = 'purple')
# ax.plot(x, y**2, label ='n**2', color = 'blue')
# ax.plot(x, y**3, label ='n**3', color = 'green')
# ax.plot(x, 2**y, label ='2**n', color = 'orange')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label ='log n', color = 'black')
ax.plot(x,y, label ='n', color = 'red')
ax.plot(x, x*y, label ='n log n', color = 'purple')
ax.plot(x, y**2, label ='n**2', color = 'blue')
# ax.plot(x, y**3, label ='n**3', color = 'green')
# ax.plot(x, 2**y, label ='2**n', color = 'orange')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label ='log n', color = 'black')
ax.plot(x,y, label ='n', color = 'red')
ax.plot(x, x*y, label ='n log n', color = 'purple')
ax.plot(x, y**2, label ='n**2', color = 'blue')
ax.plot(x, y**3, label ='n**3', color = 'green')
# ax.plot(x, 2**y, label ='2**n', color = 'orange')
ax.legend()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label ='log n', color = 'black')
ax.plot(x,y, label ='n', color = 'red')
ax.plot(x, x*y, label ='n log n', color = 'purple')
ax.plot(x, y**2, label ='n**2', color = 'blue')
ax.plot(x, y**3, label ='n**3', color = 'green')
ax.plot(x, 2**y, label ='2**n', color = 'orange')
ax.legend()
```

```python
plt.subplot(3,2,1)
plt.plot(x, x)

plt.subplot(3,2,2)
plt.plot(x, y)

plt.subplot(3,2,3)
plt.plot(x, x*y)

plt.subplot(3,2,4)
plt.plot(x, y**2)

plt.subplot(3,2,5)
plt.plot(x, y**3)

plt.subplot(3,2,6)
plt.plot(x, 2**y)
```
